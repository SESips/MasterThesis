```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the bar chart
scenarios = ['No Mistakes', 'One Repeat', 'Two Repeats', 'Three Repeats']
experienced_times = [5, 10, 15, 20]  # Times for experienced user (5 mins per repeat)
inexperienced_times = [10, 20, 30, 40]  # Times for inexperienced user (10 mins per repeat)
automated_times = [0.1, 0.1, 0.1, 0.1]  # Automated time is 0 for all scenarios
```

```python
# Number of bars
n = len(scenarios)

# Bar width
bar_width = 0.2

# X locations for the groups
indices = np.arange(n)

# Create the figure and axis
fig, ax = plt.subplots()

# Plot bars for experienced times
bar1 = ax.bar(indices - bar_width, experienced_times, bar_width, label='Manual - Experienced User')

# Plot bars for inexperienced times
bar2 = ax.bar(indices, inexperienced_times, bar_width, label='Manual - Inexperienced User')

# Plot bars for automated times
bar3 = ax.bar(indices + bar_width, automated_times, bar_width, label='Automated')

# Add labels, title, and legend
ax.set_xlabel('Scenarios')
ax.set_ylabel('Time (minutes)')
ax.set_title('Process Time: Automated vs Manual for Experienced and Inexperienced Users')
ax.set_xticks(indices)
ax.set_xticklabels(scenarios)
ax.legend()

# Display the plot
plt.show()
```

### Intervention 1 quantitative evaluation

```python
# Data from the table
categories = ['Failed checks', 'Passed checks', 'NA flags', 'Total results']
before_values = [8, 1158, 2987, 4153]
after_values = [8, 1041, 287, 1336]
percentage_difference = [0, -10.10, -90.39, -67.83]
```

```python
# Plotting the line chart
plt.figure(figsize=(10, 6))

# Plot lines for before and after values
plt.plot(categories, before_values, marker='o', linestyle='-', color='blue', label='Before')
plt.plot(categories, after_values, marker='o', linestyle='-', color='green', label='After')

# Adding percentage difference as text on the graph
for i, category in enumerate(categories):
    plt.text(category, after_values[i], f'{percentage_difference[i]:.2f}%', ha='left', va='bottom', color='red')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Comparison of DQA Metrics Before and After Intervention')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
